looking for the bestseller book on 01 Dec 2020 in The New York times

```python
from requests import get

url = "https://api.nytimes.com/svc/books/v3/lists/2020-01-12/hardcover-nonfiction.json?api-key=kimMoorm10fGoBo2FlDPt2rNovh2aB45"

response = get(url)
response.status_code
```

```python
book = response.json()['results']['books']
```

```python
rank = []
title = []
author = []
contributor = []
publisher = []
book_image = []

for i in range(0,15):
    rank.append(book[i]['rank'])
    title.append(book[i]['title'])
    author.append(book[i]['author'])
    contributor.append(book[i]['contributor'])
    publisher.append(book[i]['publisher'])
    book_image.append(book[i]["book_image"])
```

```python
import pandas as pd

df = pd.DataFrame({
    "rank": rank,
    "title": title,
    "author": author,
    "contributor": contributor,
    "publisher": publisher,
    "book_image": book_image
})
```
