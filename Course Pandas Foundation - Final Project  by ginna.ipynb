```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
##change data date
df["Order Date"] = pd.to_datetime(df['Order Date'])
##filter date only in 2017
df17 = df[(df["Order Date"] >= '2017-01-01') & (df["Order Date"] <= "2017-12-31")]
##result of sum mean std
df17["Sales"].agg(["sum","mean","std"]).round(decimals=2)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
#filter date
pp19 = df[(df["Order Date"] >= '2019-01-01') & (df["Order Date"] <= "2019-12-31")]
##filter only West and central and sum
# total is 334909.5525
sum_WC_sale = pp19.query("Region == ['West','Central']")["Sales"].sum()
#sum Total Sales = 609205.598
sum_sale = pp19["Sales"].sum()


pps = (sum_WC_sale/sum_sale)*100
pps
# proportion of total sale in West+Central = 54.97%


```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
#filter date during 2019-2020
df20 = df[(df["Order Date"] >= '2019-01-01') & (df["Order Date"] <= "2020-12-31")]

#select top 10 by sort product and show price to compare
df20 =  df20.groupby("Product Name")["Sales"].agg(["count","sum"])\
        .sort_values("count",ascending=False)\
        .head(10).round(decimals=2)

df20
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
#which Region is the most sale in 2019
rgs = df[(df["Order Date"] >= '2019-01-01') & (df["Order Date"] <= "2019-12-31")]
rgs = rgs.groupby(["Region"])["Sales"].sum()
rgs.plot(kind="bar",color="salmon");
```

```python
#sale trend in 2018
df["month"] = pd.to_datetime(df['Order Date']).dt.month
df18 = df[(df["Order Date"] >= '2018-01-01') & (df["Order Date"] <= '2018-12-31')]
ss18 = df18.groupby(["month"])["Sales"].sum()

ss18.plot(kind="line");
```

```python
## line chart for total sale between 2018 and 2019
#sale tread in 2019
df19 = df[(df["Order Date"] >= '2019-01-01') & (df["Order Date"] <= "2019-12-31")]
ss19 = df19.groupby(["month"])["Sales"].sum()

#create new dataframe
gs18 = pd.DataFrame(ss18)
#change columns name
gs18.columns = ["sale 2018"]

#create new dataframe        
gs19 = pd.DataFrame(ss19)
#change columns name
gs19.columns = ["sale 2019"]
#merge dataframe
gs = pd.merge(gs18,gs19, on="month")

gs.plot(kind="line");
```
